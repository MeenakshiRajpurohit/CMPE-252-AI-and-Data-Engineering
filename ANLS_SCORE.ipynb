```python
import Levenshtein
import numpy as np

def calculate_anls(prediction, ground_truth, threshold=0.5):
    # Standardize both strings
    pred = str(prediction).lower().strip()
    gt = str(ground_truth).lower().strip()

    if not gt:
        return 0.0

    # Calculate Levenshtein distance
    distance = Levenshtein.distance(pred, gt)

    # Normalize: 1 - (distance / max_length)
    max_len = max(len(pred), len(gt))
    nls = 1 - (distance / max_len) if max_len > 0 else 1.0

    # Apply DocVQA threshold: if similarity < 0.5, score is 0
    return nls if nls >= threshold else 0.0

# Placeholder for results. You should replace this with your actual prediction and ground truth data.
results = [
    {'pred': 'hello world', 'truth': 'hello world'},
    {'pred': 'hi there', 'truth': 'hello there'},
    {'pred': 'apple', 'truth': 'aple'},
    {'pred': 'banana', 'truth': 'banna'},
    {'pred': 'cat', 'truth': 'dog'}
]

# Placeholder for split_ds. This would typically come from your dataset loading.
split_ds = {
    'test': [
        {'question': 'What is the greeting?'},
        {'question': 'What is another greeting?'},
        {'question': 'What fruit is this?'},
        {'question': 'What other fruit is this?'},
        {'question': 'What animal is this?'}
    ]
}

# 1. Run the evaluation on your results
anls_scores = [calculate_anls(r['pred'], r['truth']) for r in results]
mean_anls = np.mean(anls_scores) * 100

print(f"📊 Accuracy Comparison:")
print(f"❌ Exact Match: 15.00%")
print(f"✅ ANLS Score:  {mean_anls:.2f}%")

# 2. See the "Saved" samples
print("\n--- 💡 Samples saved by ANLS (previously marked 0%): ---")
for i, r in enumerate(results):
    score = anls_scores[i]
    # Highlight samples where EM was 0 but ANLS is high
    if r['pred'] != r['truth'] and score > 0.7:
        print(f"Q: {split_ds['test'][i]['question']}")
        print(f"  Truth: '{r['truth']}'")
        print(f"  Pred:  '{r['pred']}'")
        print(f"  ANLS:  {score:.2f}")
        print("-" * 10)
```
